```python
from selenium import webdriver
import selenium
import time
from bs4 import BeautifulSoup
import requests
```

```python
def get_seattle_urls(main_url='https://hotpads.com/seattle-wa/apartments-for-rent'):
    seattle_page_urls = []
    for page_num in range(1, 105):
        if page_num == 1:
            seattle_page_urls.append(main_url)
        else:
            seattle_page_urls.append(main_url + f'?page={page_num}')
    return seattle_page_urls
```

```python
browser = webdriver.Firefox()
```

```python
def get_listing_urls(seattle_page_urls):
    full_links = []
    for page in seattle_page_urls:
        potential_links = []
        browser.get(page)
        time.sleep(20)
        current_page_source = browser.page_source        
        soup = BeautifulSoup(current_page_source, 'html.parser')
        for link in soup.find_all('a', href=True):
            potential_links.append(link['href'])
        for each_link in potential_links:
            if each_link.endswith('pad'):
                 full_links.append('https://hotpads.com' + each_link)
        time.sleep(randint(60,120))
    return set(full_links)
```

```python
seattle_page_urls = get_seattle_urls()
```

```python
all_the_links = get_listing_urls(seattle_page_urls)
```
