```python
links = []
```

```python
for link in soup.find_all('a', href=True):
    links.append(link['href'])
```

```python
full_links = []

for each_link in links:
    if each_link.endswith('pad') and each_link not in full_links:
        full_links.append('https://hotpads.com' + each_link)
```
